```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from astropy.table import Table
from IPython.display import set_matplotlib_formats
set_matplotlib_formats('pdf')
```

```python
t = Table.read('/home/aalvarez/Downloads/SC8R6/output_mer_cat.fits')
```

```python
fluxes = ['DETECTION_TOTAL', 'Y_TOTAL', 'J_TOTAL', 'H_TOTAL'] + [f'{b}_EXT_LSST_TOTAL' for b in 'GRIZ']
```

```python
fig, axes = plt.subplots(ncols=3, nrows=3, figsize=(20, 20))
for ax, f in zip(axes.ravel(), fluxes):
    fl, er = t[f'FLUX_{f}'], t[f'FLUXERR_{f}']
    mask = (fl > 0)
    ax.scatter(fl[mask], er[mask], marker='.', alpha=0.2, rasterized=True)
    ax.set_xlabel(f'FLUX_{f}')
    ax.set_ylabel(f'FLUXERR_{f}')
    #ax.set_yscale('log')
    ax.set_xscale('log')
plt.tight_layout()
```
