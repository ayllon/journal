```python
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt
from astropy.io import fits
from scipy import interpolate
```

```python
t=fits.open('/home/aalvarez/Work/Data/coupon/ReferenceSampleEastPhotometryShifted.fits')
shifts = eval(t['r'].header['SHIFTS'])
photo = t['NNPZ_PHOTOMETRY']
```

```python
plt.figure(figsize=(8,8))
plt.plot(shifts, photo.columns['r_shifted'].array[30], label='r')
plt.ylabel('Flux')
plt.xlabel('Shift')
_ = plt.legend()
```

```python
f=interpolate.interp1d(shifts, photo.columns['r_shifted'].array[30])
```

```python
plt.figure(figsize=(8,8))
new_range=range(min(shifts), max(shifts), 2)
plt.plot(shifts, photo.columns['r_shifted'].array[30], label='r')
plt.plot(new_range, f(new_range), 'x')
plt.plot([0], [photo.columns['r'].array[30]], 'o')
```
