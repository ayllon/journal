```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from astropy.table import Table
```

```python
bench_data = Table.read('/tmp/benchmark.csv')
repeat = 50
```

```python
source_size = 40
fixed_source = bench_data[bench_data['Source'] == source_size]
plt.figure(figsize=(20,10))
for scale in np.unique(fixed_source['Scale']):
    subset = fixed_source[fixed_source['Scale'] == scale]
    plt.plot(subset['Target'], subset['Time']/1e6/repeat, label=scale)
plt.legend()
plt.xlabel('Target size')
plt.ylabel('Time (ms)')
_ = plt.title(f'Source size {source_size}')
```

```python
target_size = 500
fixed_target = bench_data[bench_data['Target'] == target_size]
plt.figure(figsize=(20,10))
for scale in np.unique(fixed_target['Scale']):
    subset = fixed_target[fixed_target['Scale'] == scale]
    plt.plot(subset['Source'], subset['Time']/1e6/repeat, label=scale)
plt.legend()
plt.xlabel('Source size')
plt.ylabel('Time (ms)')
_ = plt.title(f'Target size {target_size}')
```
