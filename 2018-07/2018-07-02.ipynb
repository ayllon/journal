```python
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import colors
from astropy.table import Table
from astropy.io import fits
```

```python
fixed_fit = fits.open('/home/aalvarez/Work/Projects/sextractorxx/tests/check_modelfitting_fixed.fits')
fixed_residual = fits.open('/home/aalvarez/Work/Projects/sextractorxx/tests/check_residual_fixed.fits')
img = fits.open('/home/aalvarez/Work/Data/sim08/sim08.fits')
```

```python
vmin, vmax = img[0].data.min(), img[0].data.max()
norm = colors.SymLogNorm(linthresh=1, vmin=vmin, vmax=vmax)

plt.figure(figsize=(9,4))
plt.subplot(1, 2, 1)
plt.imshow(
    img[0].data,
    norm=norm,
    cmap=plt.get_cmap('binary_r'),
    vmin=vmin, vmax=vmax
)
plt.subplot(1, 2, 2)
plt.imshow(
    fixed_fit[0].data,
    alpha=0.8,
    norm=norm,
    cmap=plt.get_cmap('binary_r'),
    vmin=vmin, vmax=vmax
)
```
