```python
def f(x, A=5, b=1):
    return (x-A)**2 + b
```

```python
plt.figure()
xs = np.linspace(-1, 20, 50)
As = np.array([2, 5, 6, 10])
Bs = np.array([0, 1, 3, 20])
e = f(xs[:,np.newaxis], As, Bs)
plt.plot(xs, e)
plt.show()
```

```python
def minimize(f, guess, max_iter=1000, xtol=1e-2, epsilon=1e-1, **kwargs):
    it = 0
    
    while it < max_iter:
        # Evaluate at guess and guess + epsilon
        ype = f(guess + epsilon, **kwargs)
        yme = f(guess - epsilon, **kwargs)

        # Compute slope
        df = (ype - yme) / 2. * epsilon

        # Move guess following the slope
        guess -= df * 0.1
        
        it += 1
    
    return guess, it
```

```python
guess = np.full(len(As), 5.)
minimize(f, guess, A=As, b=Bs)
```
