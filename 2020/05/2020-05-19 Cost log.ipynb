```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from scipy.interpolate import interp1d

plt.rcParams['figure.figsize'] = (20, 10)
```

```python
def lognorm(x, a=0.2):
    return np.exp(-np.log(x)**2/a**2)

def uniform(x):
    return 1
```

```python
a = 10.** np.linspace(-5, 1, 400)
```

```python
plt.figure()
plt.plot(a, lognorm(a))
```

```python
lognorm_interp = interp1d(a, lognorm(a))
```

```python
plt.figure()
plt.plot(a, lognorm_interp(a))
```

```python
%timeit lognorm(5)
```

```python
%timeit lognorm_interp(5)
```

```python
%timeit uniform(5)
```
