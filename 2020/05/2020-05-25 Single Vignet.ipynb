```python
%matplotlib inline
from astropy.table import Table
import numpy as np
import matplotlib.pyplot as plt

plt.rcParamsDefault['figure.figsize'] = (20, 10)
```

```python
t = Table.read('/home/aalvarez/Work/Data/sim11/sim11_cat.fits')
```

```python
t['vignet'][0].shape
```

```python
idx = np.arange(len(t))
```

```python
np.random.shuffle(idx)

ncols = 5
nrows = 5

fig, axes = plt.subplots(ncols = ncols, nrows = nrows, figsize=(20, 20))

axes = axes.ravel()

for i, ax in enumerate(axes):
    obj = t[idx[i]]
    ax.imshow(obj['vignet'])
    ax.set_xticks([])
    ax.set_yticks([])
    
plt.tight_layout()
```
