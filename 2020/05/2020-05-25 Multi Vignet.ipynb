```python
%matplotlib inline
from astropy.table import Table
import numpy as np
import matplotlib.pyplot as plt

plt.rcParamsDefault['figure.figsize'] = (20, 10)
```

```python
t = Table.read('/home/aalvarez/Work/Data/sim11/sim11_cat_multi.fits')
```

```python
t['vignet'][0].shape
```

```python
n_vignets = t['vignet'][0].shape[0]
```

```python
idx = np.arange(len(t))
```

```python
np.random.shuffle(idx)

ncols = n_vignets
nrows = 10

im_size = 20 // n_vignets

fig, axes = plt.subplots(ncols = ncols, nrows = nrows, figsize=(20, im_size*nrows))

for i in range(nrows):
    obj = t[idx[i]]
    for n in range(ncols):
        axes[i, n].imshow(obj['vignet'][n])
        axes[i, n].set_xticks([])
        axes[i, n].set_yticks([])
        
plt.tight_layout()
```
