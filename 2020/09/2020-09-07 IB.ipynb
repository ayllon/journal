```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from glob import glob
import os

from IPython.display import set_matplotlib_formats
set_matplotlib_formats('pdf')
```

```python
ib = sorted(glob('/home/aalvarez/Phosphoros/AuxiliaryData/Filters/Subaru/*.dat'))
```

```python
euclid = sorted(glob('/home/aalvarez/Phosphoros/AuxiliaryData/Filters/euclid/*.dat'))
lsst = sorted(glob('/home/aalvarez/Phosphoros/AuxiliaryData/Filters/lsst/*.dat'))
```

```python
plt.figure(figsize=(20, 10))
for e in euclid + lsst:
    name = os.path.splitext(os.path.basename(e))[0]
    data = np.loadtxt(e)
    plt.plot(data[:,0], data[:,1], c='gray', linestyle=':')

cmap = plt.get_cmap('rainbow')
cs = cmap(np.arange(0, len(ib))/len(ib))
for i, c in zip(ib, cs):
    name = os.path.splitext(os.path.basename(i))[0]
    data = np.loadtxt(i)
    plt.plot(data[:,0], data[:,1], c=c, label=name)
    
plt.ylabel('Transmission')
plt.xlabel('$\lambda$')
plt.legend()
plt.tight_layout()
```
