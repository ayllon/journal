```python
%matplotlib inline
import matplotlib.pyplot as plt
from astropy.table import Table
import numpy as np
from matplotlib.colors import SymLogNorm
```

```python
catalog = Table.read('/home/aalvarez/Work/Data/sim12/sim12_cat.fits')
```

```python
idx = np.random.choice(len(catalog), 4)

fig, axes = plt.subplots(nrows=len(idx), ncols=2, figsize=(20, 40))
for i, ax in zip(idx, axes):
    ax[0].set_title(catalog['source_id'][i])
    ax[0].imshow(catalog['vignet'][i], cmap='Greys_r', norm=SymLogNorm(1000))
    ax[1].imshow(catalog['torch-jit-export'][i,0], cmap='Greys_r', norm=SymLogNorm(100))
    
plt.tight_layout()
```
