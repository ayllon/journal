```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from astropy.table import Table

plt.rcParams['figure.figsize'] = (20, 10)
```

```python
t = Table.read('/home/aalvarez/Work/Projects/SourceXtractor-litmus/data/sim11/ref/sim11_r_01_reference.fits')
```

```python
growth_nsteps = t['FLUX_GROWTH'].shape[1]
print(growth_nsteps)
```

```python
idx = [64, 65, 120, 90, 20]
```

```python
fig, axes = plt.subplots(nrows = len(idx), ncols=2, figsize=(20,30))
for i, ax in zip(idx, axes):
    obj = t[i]
    growth_px = np.arange(1, growth_nsteps + 1) * obj['FLUX_GROWTHSTEP']
    ax[0].imshow(obj['VIGNET'])
    ax[1].plot(growth_px, obj['FLUX_GROWTH']/obj['FLUX_GROWTH'][-1], 'o-')
    ax[1].axvline(obj['FLUX_RADIUS'], linestyle='--', c='black')
    ax[1].axhline(0.5, linestyle=':', c='gray')
    
plt.tight_layout()
```
