```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from astropy.table import Table
```

```python
tu = Table.read('/var/run/media/aalvarez/Astrodata/Work/Data/SC456/spread/TU.fits')
```

```python
tu.colnames
```

```python
nbins = 20
```

```python
vis_log = np.log(tu['TU_FNU_VIS'])
vis_density, vis_edges = np.histogram(vis_log, bins=nbins, density=True)
vis_density /= np.sum(vis_density)
```

```python
i_log = np.log(tu['TU_FNU_I_LSST'])
_, i_bins = np.histogram(i_log, bins=nbins)
_, z_bins = np.histogram(tu['Z_OBS'], bins=4)
```

```python
for vis_start, vis_end, vis_prob in zip(vis_edges[:-1], vis_edges[1:], vis_density):
    if vis_prob > 0.05 or vis_prob == 0.:
        print(f'{vis_start:.2f} <= log(VIS) < {vis_end:.2f} | {vis_prob:.2f}')
        
        vis_mask = np.logical_and(vis_start <= vis_log, vis_log < vis_end)
        
        i_density, i_edges = np.histogram(i_log[vis_mask], bins=i_bins, density=True)
        i_density /= np.sum(i_density)
        
        for i_start, i_end, i_prob in zip(i_edges[:-1], i_edges[1:], i_density):
            if i_prob > 0.05 or i_prob == 0.:
                print(f'\t{i_start:.2f} <= I < {i_end:.2f} | {i_prob:.2f}')
                
                i_mask = np.logical_and(i_start <= i_log, i_log < i_end)
                full_mask = np.logical_and(i_mask, vis_mask)
                
                z_density, z_edges = np.histogram(tu['Z_OBS'][full_mask], bins=z_bins, density=True)
                z_density /= np.sum(z_density)
                
                for z_start, z_end, z_prob in zip(z_edges[:-1], z_edges[1:], z_density):
                    if z_prob > 0.02 or z_prob == 0.:
                        print(f'\t\t{z_start:.2f} <= Z < {z_end:.2f} | {z_prob:.2f}')
```
