```python
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d
from astropy.table import Table
```

```python
euclidean = Table.read('/tmp/nnpz_compare/r_shift9/n_kdtree_3e3b49ed617d70d9cae119aaf2a04a63.fits')
```

```python
plt.figure(1)
for r in euclidean:
    plt.plot(np.arange(1,1001), np.cumsum(-np.sort(-r['NeighborWeights']) / np.sum(r['NeighborWeights'])), 'x-')
```

```python
w_sum = np.sum(euclidean['NeighborWeights'], axis=1)
w_norm = np.flip(np.sort(euclidean['NeighborWeights'] / w_sum[:, None]), axis=1)
```

```python
w_cumsum = np.cumsum(w_norm, axis=1)
```

```python
plt.figure()
_ = plt.boxplot(w_cumsum[:,0:100])
```
