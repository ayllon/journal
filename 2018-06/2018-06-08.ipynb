```python
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d
from astropy.table import Table
```

```python
chi2 = Table.read('/tmp/nnpz_compare/r_shift9/n_real_ac7e629e362cdda7b479c74222ad2c81.fits')
```

```python
def plot_distance_vs_weight(distances, weights, marker=None):
    sorted_index = np.flip(distances.argsort(), axis=0)
    plt.plot(distances[sorted_index], weights[sorted_index], marker=marker)
```

```python
plt.figure()
#plot_distance_vs_weight(chi2['NeighborDistances'].flatten(), chi2['NeighborWeights'].flatten())
plot_distance_vs_weight(chi2['NeighborDistances'][4], chi2['NeighborWeights'][4], 'x')
plot_distance_vs_weight(chi2['NeighborDistances'][0], chi2['NeighborWeights'][0], '.')
plot_distance_vs_weight(chi2['NeighborDistances'][1760], chi2['NeighborWeights'][1760], '^')
plot_distance_vs_weight(chi2['NeighborDistances'][3], chi2['NeighborWeights'][3], '<')
plot_distance_vs_weight(chi2['NeighborDistances'][2], chi2['NeighborWeights'][2], '>')

plt.xlabel('$Distance\ \chi^2$')
plt.ylabel('$Weight\ likelihood$')
```
